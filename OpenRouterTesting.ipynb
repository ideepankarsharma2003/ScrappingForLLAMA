# OpenRouter

```python
from basic_cleaner import clean

original_content_of_website= clean(url)
original_content_of_website
```

```python
from summa import summarizer
trimmed_content= summarizer.summarize(original_content_of_website, words=500)
trimmed_content= trimmed_content.replace('\n', '')
trimmed_content
```

```python
response = openai.ChatCompletion.create(
        model="openai/gpt-3.5-turbo", # Optional (user controls the default)
        messages=[
                    {"role": "user", "content": f"Can you suggest organic keywords for high seo ranking from the following content: \n{trimmed_content}"},
                ],
            # headers={ "HTTP-Referer": $YOUR_SITE_URL, # To identify your app
            #         "X-Title": $YOUR_APP_NAME },
            headers={ "HTTP-Referer": "https://openrouter.ai/api/v1", # To identify your app
                    "X-Title": "Deepankar" },
)
reply = response.choices[0].message
```
